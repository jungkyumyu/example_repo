```python
beerqa_dir = '/home/jk/Jupyter_works/BeerQA/beerqa_train_v1.0.json'

with open(beerqa_dir) as file:
    BeerQA_jsn = json.load(file)
    
```

```python
BeerQA_data = BeerQA_jsn['data']
print(len(BeerQA_data))
```

```python
rand_inx = random.randint(0, len(BeerQA_data)-1)
print(len(BeerQA_data[rand_inx]['context']))
BeerQA_data[rand_inx]
```

```python
max_hop = 0
idx_list = []

for idx, q in enumerate(BeerQA_data):
    
    context_no = len(q['context'])
    
    if context_no == max_hop:
        idx_list.append(idx)
    
    if context_no > max_hop:
        max_hop = context_no
        idx_list = []
        idx_list.append(idx)
        
print(max_hop)
print(idx_list)
```

```python
rand_idx_4 = idx_list[random.randint(0, len(idx_list))]
print(len(BeerQA_data[rand_idx_4]['context']))
BeerQA_data[rand_idx_4]
```
